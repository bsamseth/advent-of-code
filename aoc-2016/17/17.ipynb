```python
from collections import deque
from hashlib import md5
```

```python
def bfs(root, puzzle_input='vkjiggvb'):
    
    queue = deque([root])
    
    while queue:
        path, (x, y) = queue.popleft()
        
        if x == 3 and y == 3:
            return path
        
        up, down, left, right = [int(i, base=16) >= 11 for i in md5((puzzle_input + path).encode('utf-8')).hexdigest()[:4]]
        
        if down and y < 3:
            queue.append((path+'D', (x, y+1)))
        if right and x < 3:
            queue.append((path+'R', (x+1, y)))
        if left and x > 0:
            queue.append((path+'L', (x-1, y)))
        if up and y > 0:
            queue.append((path+'U', (x, y-1)))
        
```

```python
path = bfs(('', (0,0)))
```

```python
def bfs_longest(root, puzzle_input='vkjiggvb'):
    
    queue = deque([root])
    
    longest = []
    
    
    while queue:
        path, (x, y) = queue.popleft()
        
        if x == 3 and y == 3:
            print(len(path), path)
            longest.append(path)
            continue
        
        up, down, left, right = [int(i, base=16) >= 11 for i in md5((puzzle_input + path).encode('utf-8')).hexdigest()[:4]]
        
        if left and x > 0:
            queue.append((path+'L', (x-1, y)))
        if up and y > 0:
            queue.append((path+'U', (x, y-1)))
        if down and y < 3:
            queue.append((path+'D', (x, y+1)))
        if right and x < 3:
            queue.append((path+'R', (x+1, y)))
     
    return longest    
```
