```python
puzzle_input = '.^^^.^.^^^^^..^^^..^..^..^^..^.^.^.^^.^^....^.^...^.^^.^^.^^..^^..^.^..^^^.^^...^...^^....^^.^^^^^^^'
n_rows = 400000
#puzzle_input = '.^^.^.^^^^'
#n_rows = 10
```

```python
rows = [[True] + [i == '.' for i in puzzle_input] + [True]]
for i in range(1, n_rows):
    new_row = []
    pre_row = rows[i-1]
    for j in range(1, len(pre_row)-1):
        if pre_row[j-1] != pre_row[j+1]:
            new_row.append(False)
        else:
            new_row.append(True)
    rows.append([True] + new_row + [True])

print(sum(sum(row) - 2 for row in rows))
```

```python
def print_rows(rows):
    for row in rows:
        for tile in row[1:-1]:
            print('.' if tile else '^', end='')
        print()
```

```python
print_rows(rows)
```
